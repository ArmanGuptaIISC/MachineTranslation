```python
import os, sys
i = 0
for file in os.listdir():
    if file.endswith('.zip'):  
        command = f'unzip "{file}" -d fulldata/"{file[:-4]}"'   
        i+=1
        ret_val = os.system(command)    
        if ret_val!=0:
            print("Exit Code:",ret_val)
            sys.exit("Error Occured")
```
